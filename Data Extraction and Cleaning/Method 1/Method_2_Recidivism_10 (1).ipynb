```python
from pdf2image import convert_from_path
import pytesseract
import os

# Define the path to your PDF file
pdf_path = "CBT metaanalysis from prison.pdf"  # Replace with your actual PDF path
output_folder = "output_images"  # Folder to store images
os.makedirs(output_folder, exist_ok=True)

# Convert PDF to images
#poppler_path = r"C:\Users\suram\Downloads\Release-24.08.0-0\poppler-24.08.0\Library\bin"
images = convert_from_path(pdf_path)

pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

ocr_text = []

for i, img in enumerate(images):
    img_path = os.path.join(output_folder, f"page_{i+1}.png")
    img.save(img_path, "PNG")

    # Extract text using OCR
    text = pytesseract.image_to_string(img)
    ocr_text.append(text)

# Combine extracted text
full_text = "\n".join(ocr_text)

# Save extracted text to a file
with open("extracted_text.txt", "w", encoding="utf-8") as f:
    f.write(full_text)

print("OCR extraction complete. Text saved to extracted_text.txt")
```

```python
import cv2
import pytesseract
import os
from pdf2image import convert_from_path

# Define paths
pdf_path = "CBT metaanalysis from prison.pdf"  # Replace with your PDF file
output_folder = "processed_images"  # Folder to store processed images
os.makedirs(output_folder, exist_ok=True)

# Convert PDF to images
#poppler_path = r"C:\Users\suram\Downloads\Release-24.08.0-0\poppler-24.08.0\Library\bin"
images = convert_from_path(pdf_path)

# Configure Tesseract
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

extracted_text = []

for i, img in enumerate(images):
    img_path = os.path.join(output_folder, f"page_{i+1}.png")
    img.save(img_path, "PNG")

    # Read the saved image using OpenCV
    image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Convert to grayscale

    # Apply Adaptive Thresholding (Improves OCR performance)
    processed_img = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 11, 2)

    # Noise removal using GaussianBlur
    processed_img = cv2.GaussianBlur(processed_img, (5, 5), 0)

    # Save processed image
    processed_img_path = os.path.join(output_folder, f"processed_page_{i+1}.png")
    cv2.imwrite(processed_img_path, processed_img)

    # Perform OCR on processed image
    text = pytesseract.image_to_string(processed_img, config='--psm 6')
    extracted_text.append(text)

# Combine extracted text and save to a file
full_text = "\n".join(extracted_text)
with open("extracted_text_improved.txt", "w", encoding="utf-8") as f:
    f.write(full_text)

print("OCR extraction complete. Text saved to extracted_text_improved.txt")
```

```python
import pandas as pd
import re
import spacy

# Load the English NLP model
nlp = spacy.load("en_core_web_sm")

# Read extracted text
with open("extracted_text_improved.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Function to clean the text
def clean_text(text):
    # Remove non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)

    # Remove multiple spaces, new lines
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove special characters (except for important ones like % and $)
    text = re.sub(r'[^\w\s.,%-]', '', text)

    # Convert text to lowercase
    text = text.lower()

    return text

# Apply text cleaning
cleaned_text = clean_text(text)

# Split into structured lines (assuming one sentence per line)
lines = cleaned_text.split(". ")

# Convert into structured format
df = pd.DataFrame({"Sentence": lines})

# Save to CSV
df.to_csv("CBT_metaanalysis_from_prison_cleaned_text.csv", index=False, encoding="utf-8")

print("Cleaned text saved to cleaned_text.csv")
```

```python
import pandas as pd
import spacy
import re

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

# Read cleaned text
with open("CBT_metaanalysis_from_prison_cleaned_text.csv", "r", encoding="utf-8") as f:
    lines = f.readlines()

# Function to extract Named Entities
def extract_entities(text):
    doc = nlp(text)
    entities = {"ORG": [], "PERSON": [], "DATE": [], "GPE": [], "MONEY": [], "PERCENT": [], "QUANTITY": []}

    for ent in doc.ents:
        if ent.label_ in entities:
            entities[ent.label_].append(ent.text)

    return entities

# Function to classify sentences
def classify_sentence(sentence):
    sentence = sentence.lower()

    if any(word in sentence for word in ["introduction", "background", "objective"]):
        return "Introduction"
    elif any(word in sentence for word in ["method", "procedure", "approach"]):
        return "Methodology"
    elif any(word in sentence for word in ["result", "analysis", "findings"]):
        return "Results"
    elif any(word in sentence for word in ["conclusion", "summary", "final"]):
        return "Conclusion"
    else:
        return "Other"

# Process each sentence
data = []
for line in lines:
    line = line.strip()
    if line:
        entities = extract_entities(line)
        category = classify_sentence(line)

        data.append({
            "Sentence": line,
            "Category": category,
            "Organizations": ", ".join(entities["ORG"]),
            "Persons": ", ".join(entities["PERSON"]),
            "Dates": ", ".join(entities["DATE"]),
            "Locations": ", ".join(entities["GPE"]),
            "Numbers": ", ".join(entities["QUANTITY"])
        })

# Convert into a DataFrame
df = pd.DataFrame(data)

# Save structured text to CSV
df.to_csv("structured_text.csv", index=False, encoding="utf-8")

print("NLP processing complete. Data saved to structured_text.csv")
```

```python
import pandas as pd
import torch
from sentence_transformers import SentenceTransformer
import chromadb

# Load pre-trained embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")  # Efficient embedding model

# Convert sentences to embeddings
df["embedding"] = df["Sentence"].apply(lambda x: model.encode(x).tolist())

# Save processed embeddings
df.to_csv("vectorized_text.csv", index=False, encoding="utf-8")

print("Embeddings generated successfully!")
```

```python
# Initialize ChromaDB client
chroma_client = chromadb.PersistentClient(path="./chroma_db")

# Create a new collection
collection = chroma_client.get_or_create_collection(name="cbt_analysis")

# Insert data into ChromaDB
for i, row in df.iterrows():
    collection.add(
        ids=[str(i)],
        embeddings=[row["embedding"]],
        metadatas=[{"sentence": row["Sentence"], "category": row["Category"]}]
    )

print("Data stored in ChromaDB successfully!")
```

```python
import os
import pandas as pd
import re
import spacy
from unidecode import unidecode
import nltk
from docling.document_converter import DocumentConverter
from google.colab import files

# Downloading necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

# Loading spaCy model for lemmatization
nlp = spacy.load("en_core_web_sm")

# Loading stopwords
stop_words = set(nltk.corpus.stopwords.words("english"))

# file path
pdf_path = 'CBT_meta_analysis_sex_offenders.pdf'

# Checking if file exists in the given path
if not os.path.exists(pdf_path):
    print("The file at" ,pdf_path, "does not exist.")
else:
    # Initializing the DocumentConverter
    converter = DocumentConverter()

    # Converting from PDF to a document object
    result = converter.convert(pdf_path)

    # Extracting document text and tables
    document = result.document
    raw_text = document.export_to_text()
    markdown_text = document.export_to_markdown()

    # Extracting tables from the document
    tables = document.tables

    # Cleaning text
    def clean_text(text):
        # NLP cleaning: removing noise, normalizing text, and structuring content

        # Unicode normalization
        text = unidecode(text)

        # Removing excessive whitespace & newlines
        text = re.sub(r'\s+', ' ', text).strip()

        # Expanding contractions (e.g., "don't" → "do not")
        contractions = {
            "n't": " not", "'re": " are", "'s": " is", "'ll": " will",
            "'d": " would", "'ve": " have", "'m": " am"
        }
        for key, value in contractions.items():
            text = text.replace(key, value)

        # Removing special characters & punctuation
        text = re.sub(r"[^a-zAZ0-9.]", " ", text)

        # Tokenizing sentences
        sentences = nltk.sent_tokenize(text)

        # Removing stopwords & perform lemmatization
        cleaned_sentences = []
        for sent in sentences:
            doc = nlp(sent.lower())
            tokens = [token.lemma_ for token in doc if token.text not in stop_words]
            cleaned_sentences.append(" ".join(tokens))

        # Reconstructing the text
        cleaned_text = " ".join(cleaned_sentences)

        # Attempting to structure content based on common section headers
        structured_text = re.sub(r'(?i)(Abstract|Introduction|Methods|Results|Discussion|Conclusion):', r'\n### \1\n', cleaned_text)

        return structured_text

    # Cleaning the extracted text
    cleaned_text = clean_text(raw_text)

    # Converting the tables into DataFrames and cleaning
    table_dfs = []
    for table in tables:
        table_df = pd.DataFrame(table)
        table_dfs.append(table_df)

    # Combining all tables into one DataFrame
    final_table_df = pd.concat(table_dfs, ignore_index=True)

    # Creating a DataFrame for text data
    text_data_df = pd.DataFrame([{"section": "full_text", "content": cleaned_text}])

    # Combining text and table data into a single CSV
    final_df = pd.concat([text_data_df, final_table_df], ignore_index=True)

    # output CSV file path
    csv_filename = pdf_path.replace(".pdf", "_cleaned_data.csv")

    # Save as CSV
    final_df.to_csv(csv_filename, index=False)
    files.download(csv_filename)

    # Print confirmation
    print("All cleaned text & table data saved in: ", csv_filename)

    # Save cleaned text as markdown file
    markdown_filename = pdf_path.replace(".pdf", "_cleaned_data.md")
    with open(markdown_filename, 'w') as markdown_file:
        markdown_file.write(markdown_text)
    files.download(markdown_filename)

    print("Cleaned text also saved as Markdown in: ", markdown_filename)
```
