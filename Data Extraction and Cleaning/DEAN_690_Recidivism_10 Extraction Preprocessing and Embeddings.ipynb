```python
import os
import cv2
import pytesseract
import re
import spacy
import pandas as pd
import nltk
from unidecode import unidecode
from pdf2image import convert_from_path
from pdf2image.exceptions import PDFPageCountError
from docling.document_converter import DocumentConverter
import camelot

# Downloading necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

# Loading spaCy model for lemmatization
nlp = spacy.load("en_core_web_sm")

# Loading stopwords
stop_words = set(nltk.corpus.stopwords.words("english"))

# Configure Tesseract
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

def ocr_extraction(pdf_name, images, output_folder):
    extracted_text = []
    for i, img in enumerate(images):
        img_path = os.path.join(output_folder, f"page_{i+1}.png")
        img.save(img_path, "PNG")

        # Read the saved image using OpenCV
        image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

        # Apply Adaptive Thresholding
        processed_img = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                              cv2.THRESH_BINARY, 11, 2)
        processed_img = cv2.GaussianBlur(processed_img, (5, 5), 0)

        # Save processed image
        processed_img_path = os.path.join(output_folder, f"processed_page_{i+1}.png")
        cv2.imwrite(processed_img_path, processed_img)

        # Perform OCR
        text = pytesseract.image_to_string(processed_img, config='--psm 6')
        extracted_text.append(text)

    return "\n".join(extracted_text)

def docling_extraction(pdf_path):
    converter = DocumentConverter()
    result = converter.convert(pdf_path)
    document = result.document
    raw_text = document.export_to_text()
    tables = document.tables

    def clean_text(text):
        text = unidecode(text)
        text = re.sub(r'\s+', ' ', text).strip()
        text = re.sub(r"[^a-zA-Z0-9. ]", "", text)
        sentences = nltk.sent_tokenize(text)
        cleaned_sentences = [" ".join([token.lemma_ for token in nlp(sent.lower()) if token.text not in stop_words]) for sent in sentences]
        return " ".join(cleaned_sentences)

    cleaned_text = clean_text(raw_text)
    table_dfs = [pd.DataFrame(table).astype(str) for table in tables]
    return cleaned_text, table_dfs

def camelot_extraction(pdf_path):
    tables = camelot.read_pdf(pdf_path, pages='all', flavor='stream')
    return [table.df.astype(str) for table in tables]

def remove_duplicates(text):
    seen_sentences = set()
    final_text = []
    for sentence in text.split(". "):
        sentence = sentence.strip()
        if sentence and sentence not in seen_sentences:
            seen_sentences.add(sentence)
            final_text.append(sentence)
    return " ".join(final_text)

def deduplicate_tables(tables):
    unique_tables = []
    seen_hashes = set()

    for table in tables:
        if isinstance(table, pd.DataFrame):
            # Convert all elements to strings to avoid unhashable type errors
            table = table.astype(str)

            # Compute a hashable representation (e.g., a tuple of tuples)
            table_hash = tuple([tuple(row) for row in table.values])

            if table_hash not in seen_hashes:
                seen_hashes.add(table_hash)
                unique_tables.append(table)

    return unique_tables

def process_pdfs(folder_path, output_base_folder):
    pdf_files = [f for f in os.listdir(folder_path) if f.endswith(".pdf") or f.endswith(".PDF")]

    for pdf_file in pdf_files:
        pdf_path = os.path.join(folder_path, pdf_file)
        pdf_name = os.path.splitext(pdf_file)[0]
        output_folder = os.path.join(output_base_folder, pdf_name)
        os.makedirs(output_folder, exist_ok=True)

        print(f"Processing: {pdf_file}")
        try:
          images = convert_from_path(pdf_path)
          ocr_text = ocr_extraction(pdf_name, images, output_folder)
          docling_text, docling_tables = docling_extraction(pdf_path)
          camelot_tables = camelot_extraction(pdf_path)

          combined_text = remove_duplicates(ocr_text + " " + docling_text)
          unique_tables = deduplicate_tables(docling_tables + camelot_tables)

          text_output_path = os.path.join(output_folder, f"{pdf_name}_text.txt")
          print(f"Successfully processed: {pdf_file}")
        except PDFPageCountError as e:
          print(f"Skipping {pdf_file} due to error: {e}")
        except Exception as e:
          print(f"Unexpected error for {pdf_file}: {e}")
        with open(text_output_path, "w", encoding="utf-8") as f:
            f.write(combined_text)

        for i, table in enumerate(unique_tables):
            table_output_path = os.path.join(output_folder, f"{pdf_name}_table_{i+1}.csv")
            table.to_csv(table_output_path, index=False)

        print(f"Processing complete for: {pdf_file}")

# Folder paths
input_folder = "drive/MyDrive/Colab Notebooks/Developer 2"  # Folder containing PDFs
output_folder = "drive/MyDrive/Colab Notebooks/processed_results"  # Folder to store results
os.makedirs(output_folder, exist_ok=True)

process_pdfs(input_folder, output_folder)
print("All PDFs processed successfully.")
```

```python
import os
import pandas as pd
import re
import spacy

# Load the English NLP model
nlp = spacy.load("en_core_web_sm")

# Define the processed results folder
processed_folder = "drive/MyDrive/Colab Notebooks/processed_results"

# Function to clean the text
def clean_text(text):
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)  # Remove non-ASCII characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove multiple spaces, new lines
    text = re.sub(r'[^\w\s.,%-]', '', text)  # Remove special characters
    return text.lower()  # Convert text to lowercase

# Function to extract Named Entities
def extract_entities(text):
    doc = nlp(text)
    entities = {"ORG": [], "PERSON": [], "DATE": [], "GPE": [], "MONEY": [], "PERCENT": [], "QUANTITY": []}
    for ent in doc.ents:
        if ent.label_ in entities:
            entities[ent.label_].append(ent.text)
    return entities

# Function to classify sentences
def classify_sentence(sentence):
    sentence = sentence.lower()
    if any(word in sentence for word in ["introduction", "background", "objective"]):
        return "Introduction"
    elif any(word in sentence for word in ["method", "procedure", "approach"]):
        return "Methodology"
    elif any(word in sentence for word in ["result", "analysis", "findings"]):
        return "Results"
    elif any(word in sentence for word in ["conclusion", "summary", "final"]):
        return "Conclusion"
    else:
        return "Other"

# Iterate through folders inside `processed_results`
for folder in os.listdir(processed_folder):
    folder_path = os.path.join(processed_folder, folder)

    if os.path.isdir(folder_path):  # Ensure it's a folder
        for file in os.listdir(folder_path):
            if file.endswith(".txt"):  # Process only .txt files
                file_path = os.path.join(folder_path, file)
                print(f"Processing: {file_path}")

                # Read extracted text
                with open(file_path, "r", encoding="utf-8") as f:
                    text = f.read()

                # Clean text
                cleaned_text = clean_text(text)

                # Split into sentences
                lines = cleaned_text.split(". ")

                # Store processed data
                data = []
                for line in lines:
                    line = line.strip()
                    if line:
                        entities = extract_entities(line)
                        category = classify_sentence(line)

                        data.append({
                            "Sentence": line,
                            "Category": category,
                            "Organizations": ", ".join(entities["ORG"]),
                            "Persons": ", ".join(entities["PERSON"]),
                            "Dates": ", ".join(entities["DATE"]),
                            "Locations": ", ".join(entities["GPE"]),
                            "Numbers": ", ".join(entities["QUANTITY"])
                        })

                # Convert into a DataFrame
                df = pd.DataFrame(data)

                output_folder = "drive/MyDrive/Colab Notebooks/structured_results"
                folder_path = os.path.join(output_folder, folder)
                os.makedirs(folder_path, exist_ok=True)
                # Save structured text to CSV
                output_csv = os.path.join(folder_path, f"{os.path.splitext(file)[0]}_structured.csv")
                df.to_csv(output_csv, index=False, encoding="utf-8")

                print(f"Processed and saved: {output_csv}")

print("NLP processing complete for all files.")
```

```python
import pandas as pd
import torch
from sentence_transformers import SentenceTransformer
import chromadb

structured_folder = "drive/MyDrive/Colab Notebooks/structured_results"

Presentation_folder = "drive/MyDrive/Colab Notebooks/Presentation_results"

for folder in os.listdir(structured_folder):
    folder_path = os.path.join(structured_folder, folder)

    if os.path.isdir(folder_path):  # Ensure it's a folder
        for file in os.listdir(folder_path):
            if file.endswith(".csv"):  # Process only .txt files
                file_path = os.path.join(folder_path, file)
                print(f"Processing: {file_path}")

                # Read extracted text
                df = pd.read_csv(file_path)

                # Load pre-trained embedding model
                model = SentenceTransformer("all-MiniLM-L6-v2")  # Efficient embedding model

                # Convert sentences to embeddings
                df["embedding"] = df["Sentence"].apply(lambda x: model.encode(x).tolist())

                output_path = os.path.join(Presentation_folder, folder)

                # Save processed embeddings
                df.to_csv(Presentation_folder, index=False, encoding="utf-8")

                print("Embeddings generated successfully!")

                # Initialize ChromaDB client
                chroma_client = chromadb.PersistentClient(path="./chroma_db")

                # Create a new collection
                collection = chroma_client.get_or_create_collection(name="cbt_analysis")

                # Insert data into ChromaDB
                for i, row in df.iterrows():
                  collection.add(
                    ids=[str(i)],
                    embeddings=[row["embedding"]],
                    metadatas=[{"sentence": row["Sentence"], "category": row["Category"]}]
                )

                print("Data stored in ChromaDB successfully!")



```
