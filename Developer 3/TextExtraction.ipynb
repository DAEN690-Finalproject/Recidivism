```python
import os
import re
import pdfplumber
import pandas as pd
from docling.document_converter import DocumentConverter
from google.colab import files

# Utility function to extract text from a TableCell string
def extract_cell_text(cell_str):
    # Find all occurrences of text='...'
    matches = re.findall(r"text='(.*?)'", cell_str)
    return " ".join(matches) if matches else cell_str

# Define the folder containing your PDFs
pdf_folder = "/content/Developer3"

# Define the main output folder for extracted data
output_folder = "/content/ExtractedData"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Recursively gather all PDF files from the pdf_folder
pdf_files = [
    os.path.join(root, file)
    for root, dirs, files in os.walk(pdf_folder)
    for file in files if file.lower().endswith('.pdf')
]

# Initialize the DocumentConverter for text extraction
converter = DocumentConverter()

for pdf_path in pdf_files:
    print("Processing:", pdf_path)
    base_filename = os.path.splitext(os.path.basename(pdf_path))[0]

    # Create a separate subfolder for this PDF inside the output folder
    pdf_output_folder = os.path.join(output_folder, base_filename)
    if not os.path.exists(pdf_output_folder):
        os.makedirs(pdf_output_folder)

    # ----- Extract and Save Text as Markdown -----
    try:
        result = converter.convert(pdf_path)
        document = result.document
        markdown_text = document.export_to_markdown()

        md_filename = os.path.join(pdf_output_folder, f"{base_filename}_extracted_text.md")
        with open(md_filename, "w") as md_file:
            md_file.write(markdown_text)
        print("Saved extracted text (Markdown) to:", md_filename)
        files.download(md_filename)
    except Exception as e:
        print("Error extracting text from", pdf_path, ":", e)

    # ----- Extract and Save Tables Using pdfplumber -----
    try:
        with pdfplumber.open(pdf_path) as pdf:
            table_count = 0
            for page_num, page in enumerate(pdf.pages, start=1):
                tables = page.extract_tables()
                if tables:
                    for t_idx, table in enumerate(tables, start=1):
                        df = pd.DataFrame(table)
                        # Clean each cell if it contains a TableCell string
                        df = df.applymap(lambda x: extract_cell_text(x) if isinstance(x, str) and "TableCell" in x else x)
                        csv_filename = os.path.join(pdf_output_folder, f"{base_filename}_page_{page_num}_table_{t_idx}.csv")
                        df.to_csv(csv_filename, index=False, header=False)
                        print("Saved table CSV:", csv_filename)
                        files.download(csv_filename)
                        table_count += 1
            if table_count == 0:
                print("No tables found in:", pdf_path)
    except Exception as e:
        print("Error extracting tables with pdfplumber from", pdf_path, ":", e)

print("Extraction process completed!")
```
